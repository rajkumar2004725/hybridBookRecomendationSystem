```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
def get_recommendations(title):
    idx = new_book.index[new_book['Book-Title'] == title].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10
    book_indices = [i[0] for i in sim_scores]
    return new_book['Book-Title'].iloc[book_indices]
```

```python
X=rating_with_book.groupby('User-ID').count()['Book-Rating']>200
```

```python
good_users =X[X].index
```

```python
filtered_ratings = rating_with_book[rating_with_book['User-ID'].isin(good_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
similarity_scores
```

```python
def colabrative_recommendation(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = df_books_for_rating[df_books_for_rating['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
colabrative_recommendation('A Bend in the Road')
```

```python
def hybridRecommendation(book_name):
  print("Hybrid Recommendation")
  print("---------------------")
  print("content based:")
  print("---------------------")
  c=get_recommendations(book_name)
  print(c)
  print("---------------------")
  print("colabrative:")
  print("---------------------")
  p=colabrative_recommendation(book_name)
  print(p)
```
